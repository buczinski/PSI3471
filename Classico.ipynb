```python
import os
import cv2
import numpy as np
from skimage.feature import hog
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
# Função para extrair HOG
def extrair_hog(img):
    # Converte para escala de cinza
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Calcula HOG
    hog_features = hog(gray, 
                        orientations=9, 
                        pixels_per_cell=(8, 8), 
                        cells_per_block=(2, 2), 
                        block_norm='L2-Hys', 
                        visualize=False, 
                        transform_sqrt=True)
    return hog_features
```

# Importando os outputs do arquivo 'alinha', lendo as imagens e extraindo o histograma de gradiente.

```python
kirmizi_reduzida = '/kaggle/input/alinha/imagens_processadas/Kirmizi_Pistachio_reduzida'
siirt_reduzida = '/kaggle/input/alinha/imagens_processadas/Siirt_Pistachio_reduzida'

# Inicializa listas para features e labels
features = []
labels = []

# Ler imagens de cada pasta
for file in os.listdir(kirmizi_reduzida):
    path = os.path.join(kirmizi_reduzida, file)
    img = cv2.imread(path)
    feat = extrair_hog(img)
    features.append(feat)
    labels.append(0)  # 0 para Kirmizi

for file in os.listdir(siirt_reduzida):
    path = os.path.join(siirt_reduzida, file)
    img = cv2.imread(path)
    feat = extrair_hog(img)
    features.append(feat)
    labels.append(1)  # 1 para Siirt
```

# Separando os dados e treinando o SVM para as imagens reduzidas

```python
# Converte para arrays
X = np.array(features)
y = np.array(labels)

# Separar treino e teste (200/1948)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 200, random_state=42, stratify=y)

# Cria e treina o SVM
svm = SVC(kernel='rbf', C=1.0)
svm.fit(X_train, y_train)

# Avalia no conjunto de teste
y_pred = svm.predict(X_test)
```

```python
scores = cross_val_score(svm, X, y, cv=10)
print("Acurácia média: %.2f%%" % (np.mean(scores)*100))
```
